```python
import pandas as pd
import re
```

```python
# 读取文件
excel_file1 = pd.ExcelFile('dish_data/my_h_dish_category.xlsx')
excel_file2 = pd.ExcelFile('dish_data/my_h_dish_info_all.xlsx')

# 获取指定工作表中的数据
df1 = excel_file1.parse('my_h_dish_category')
df2 = excel_file2.parse('my_h_dish_info_all')
```

```python
# 删除dish_name为空的数据行和没用的数据列
df2 = df2.dropna(subset=['dish_name'])
df2 = df2.drop(columns=['img_url', 'img_path'])
# 查看数据基本信息
print("df1形状:", df1.shape)
print("df2形状:", df2.shape)
```

```python
# 合并df2的macronutrients、vitamin和minerals列，以\n的形式拼接
df2['all_combined'] = df2['macronutrients'].fillna('') + '\n' + df2['vitamin'].fillna('') + '\n' + df2['minerals'].fillna('')

# 修改爬取过程中忽略的参数（维生素B6、维生素B12---维生素B₆、维生素B₁₂）
def process_text(text):
    text = text.replace('维生素B6', '维生素B₆')
    text = text.replace('维生素B12', '维生素B₁₂')
    return text
df2['all_combined'] = df2['all_combined'].apply(process_text)
```

```python
df2['all_combined'].head().tolist()
```

```python
# 专门用于提取分类时去除NRV和单位的函数
all_categories = []
def clean_for_category(text):
    # 去除NRV的百分比和克数等内容
    cleaned = re.sub(r'\d+% NRV|\d+\.?\d* (μg RAE|mg α-TE|mg|μg|kcal|g)', '', text)
    return cleaned.strip()
# 提取所有分类
for item in df2['all_combined']:
    categories = item.split('\n')
    for category in categories:
        # 对每个分类单独清理，不影响原始列
        cleaned_category = clean_for_category(category)
        if cleaned_category and cleaned_category not in all_categories:
            all_categories.append(cleaned_category)

print('所有分类：', all_categories)
```

```python
# 提取出所有分类如下
cn_to_en = {
    "能量": "energy",
    "蛋白质": "protein",
    "脂肪": "fat",
    "碳水化合物": "carbohydrates",
    "维生素A": "vitamin_A",
    "维生素E": "vitamin_E",
    "硫胺素": "thiamine",
    "核黄素": "riboflavin",
    "维生素B₆": "vitamin_B₆",
    "维生素B₁₂": "vitamin_B₁₂",
    "烟酸": "niacin",
    "叶酸": "folic_acid",
    "维生素C": "vitamin_C",
    "生物素": "biotin",
    "总胆碱": "total_choline",
    "维生素D": "vitamin_D",
    "维生素K": "vitamin_K",
    "泛酸": "pantothenic_acid",
    "钠": "sodium",
    "钾": "potassium",
    "镁": "magnesium",
    "铁": "iron",
    "锌": "zinc",
    "钙": "calcium",
    "磷": "phosphorus",
    "硒": "selenium",
    "碘": "iodine",
    "铜": "copper",
    "锰": "manganese"
}
len(cn_to_en)
```

```python
# 遍历每行数据处理
for idx, row in df2.iterrows():
    # 获取原始文本（如无则跳过）
    combined_text = row.get('all_combined', '')
    if not combined_text:
        continue
    
    # 按换行符拆分每个营养素条目
    elements = combined_text.split('\n')
    for elem in elements:
        if not elem:
            continue  # 跳过空行
        
        # 匹配中文名称（优先长名称，避免误判）
        matched_cn = None
        for cn in sorted(cn_to_en.keys(), key=len, reverse=True):
            if elem.startswith(cn):
                matched_cn = cn
                break
        if not matched_cn:
            continue  # 跳过未匹配的条目
        
        # 获取对应的英文名称
        en = cn_to_en[matched_cn]
        
        # 正则表达式：提取nrv百分比、数值、单位信息
        pattern = re.compile(
            fr'{re.escape(matched_cn)}'  # 匹配中文名称（如"维生素A"）
            r'(\d+)% NRV'  # 提取nrv百分比数值（如"16%"中的"16"）
            r'(\d+\.?\d*)\s*'  # 提取含量数值（如"126.00"）
            r'(.*)'  # 提取完整单位（如"μg RAE"、"mg α-TE"、"kcal"）
        )
        match = pattern.search(elem)
        
        if match:
            # 提取匹配结果
            nrv_percent = match.group(1)  # nrv百分比数值
            num_value = match.group(2)     # 含量数值
            unit = match.group(3).strip()  # 含量单位（去除首尾空格）
            nrv_unit = "% NRV"             # 固定nrv单位
            
            # 定义四列列名
            nrv_percent_col = f'{en}_nrv_percent'  # 如"energy_nrv_percent"
            nrv_unit_col = f'{en}_nrv_unit'        # 如"energy_nrv_unit"
            num_col = f'{en}_num'                  # 如"energy_num"
            unit_col = f'{en}_unit'                # 如"energy_unit"
            
            # 确保列存在（首次出现时创建，指定数据类型）
            for col, dtype in [
                (nrv_percent_col, 'Int64'),    # 整数类型，支持缺失值
                (nrv_unit_col, 'object'),      # 字符串类型
                (num_col, 'float64'),          # 浮点类型
                (unit_col, 'object')           # 字符串类型
            ]:
                if col not in df2.columns:
                    df2[col] = pd.Series(dtype=dtype)
            
            # 填充数据（处理空值）
            df2.at[idx, nrv_percent_col] = int(nrv_percent) if nrv_percent else pd.NA
            df2.at[idx, nrv_unit_col] = nrv_unit
            df2.at[idx, num_col] = float(num_value) if num_value else pd.NA
            df2.at[idx, unit_col] = unit if unit else pd.NA
```

```python
# 提取能量列纯卡路里部分
df1['pure_calorie'] = df1['calorie'].str.extract(r'(\d+\.?\d*)\s*kcal').astype(float)
df1['pure_calorie'].head()
```

```python
# 2. 用左连接替代循环匹配，高效关联分类信息
# 基于'dish_name'和'pure_calorie'两列匹配，保留df2的所有行
df2 = df2.merge(
    df1[['dish_name', 'pure_calorie', 'category']],  # 只取需要的列
    left_on=['dish_name', 'energy_num'],          # df2的匹配列
    right_on=['dish_name', 'pure_calorie'],           # df1的匹配列
    how='left'                                        # 左连接：保留df2所有行
)

# 3. 处理未匹配到的情况（填充为'/'），并删除临时列
df2['category'] = df2['category'].fillna('/')  # 未匹配到的分类用'/'表示
```

```python
def extract_num_unit(text):
    if pd.isna(text):  # 先处理NaN值，避免后续报错
        return ""
    # 正则表达式：匹配“数字（含小数）+ 单位”，支持中文单位（克/毫升/升）和英文单位（mg/μg/ml/L）
    # [克毫升升mgμgmlL] → 匹配常见单位（不区分大小写，通过flags=re.IGNORECASE实现）
    match = re.search(r'(\d+\.?\d*)\s*[克毫升升mgμgmlL]', str(text), flags=re.IGNORECASE)
    return match.group(1).strip() if match else ""  # 有匹配则返回结果，无匹配返回空
# 应用函数处理列
df2['measurement_unit'] = df2['measurement_unit'].apply(extract_num_unit)
# 先将NaN填充为空字符串，再替换指定无效值
df2['quantity'] = df2['quantity'].fillna('').str.replace(r'.*未获取到单位量.*','',regex=True)
df2['cooking_method'] = df2['cooking_method'].fillna('').str.replace(r'.*未获取到数据\n.*','',regex=True)
```
